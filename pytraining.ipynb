```python
import webbrowser
import speech_recognition as sr

r = sr.Recognizer()
r.energy_threshold = 5000
with sr.Microphone() as source:
    print("Please speak the website name loudly....")
    audio = r.listen(source)
    try:
        text = r.recognize_google(audio)
        print("You said: {}".format(text))
        text_for_url = '+'.join(text.split())
        url = 'https://www.' + text_for_url + '.com'
        webbrowser.open(url)
    except:
        print("Can't Recognize your voice please")
```

```python
import pyautogui
import tkinter as tk
from tkinter import messagebox
 
# Function to take screenshot
def take_screenshot():
    win.withdraw()
    win.after(500, capture_and_save_screenshot)
def capture_and_save_screenshot():
    screenshot = pyautogui.screenshot()
    screenshot.save("screenshot.png")
    win.deiconify()
    messagebox.showinfo("Screenshot", "Screenshot saved as screenshot.png")
 
 
win = tk.Tk()
win.title("Screenshot Taker")
win.geometry("300x150")
 
# Create a button to take screenshot
btn_screenshot = tk.Button(win,text="Take Screenshot",font=('Times New Roman', 20),bg = '#87CEEB',command = take_screenshot)
btn_screenshot.pack(pady=20)
win.mainloop()
```
